```python
import torchvision
from torch.utils.data import Dataset
from PIL import Image
import numpy as np
import glob
```

```python
def is_valid_image_file(filename):
  # Check file name extension
  valid_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff']
  if os.path.splitext(filename)[1].lower() not in valid_extensions:
    print(f"Invalid image file extension \"{filename}\". Skipping this file...")
  # Verify that image file is intact
  try:
    with Image.open(filename) as img:
      img.verify()  # Verify if it's an image
      return True
  except (IOError, SyntaxError) as e:
    print(f"Invalid image file {filename}: {e}")
    return False
```

```python
class biased_cars(Dataset):
  def __init__(self, img_dir, transform=None, target_transform=None):
    self.img_dir = "/content/drive/MyDrive/NB240/Assignment2/biased_cars_1/biased_cars_1"
    self.transform = transform
    self.target_transform = target_transform

    # Define class dictionary; classes will be training vs. validation
    self.class_dict = {"train": 0, "val": 1} # Changed to map folder names to numeric classes

    # Create empty dict for assigning label
    image_label_dict = {}

    # Initialize class counts
    class_counts = {"train": 0, "val": 0}

    # Define logic to load files
    for filename in glob.glob(self.img_dir + "/**/*.png", recursive=True): # use glob.glob to find all possible files in overall data dir
      if is_valid_image_file(filename): # Filename already includes path

      # Get base filename
        base_filename = os.path.basename(filename)

        if base_filename in att_dict: # Check to see if file is in the dictionary

          # Fill image label dict with class info. Filename will include paths (easier for loading later on)
          image_label_dict[filename] = att_dict[base_filename]

          # Obtain class by getting grandparent folder name of file
          img_class = os.path.basename(os.path.dirname(os.path.dirname(filename))) # define class

          # Map class to dictionary and count only if it is 'train' or 'val'
          if img_class in class_counts: # Check if img_class is 'train' or 'val'
            class_counts[img_class] += 1  # Increment count for valid class
            print("Image loaded:", filename)
          else:
            print(f"Skipping image in folder: {img_class}, not a train/val folder")

        else:
          print("Filename not found in attribute dictionary:", filename, ". Skipping this file...")

    # self.items is a list of tuples like: [ ("im1.jpg", 0), ("img2.png", 1), ... ]
    self.items = list(image_label_dict.items())
   #print("Class counts:", class_counts)

  def __len__(self):
    return len(self.items)

  def __getitem__(self, idx):
    img_path = self.items[idx][0]
    # image = torchvision.io.read_image(img_path) # This version reads the image directly as a Tensor
    image = Image.open(img_path)
    label = self.items[idx][1]

    if self.transform:
        image = self.transform(image)
    if self.target_transform:
        label = self.target_transform(label)

    return image, label
```
